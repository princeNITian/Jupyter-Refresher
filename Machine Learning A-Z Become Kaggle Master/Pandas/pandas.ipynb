# Merging Two DataFrame

```python
df1 = pd.DataFrame([['a',1],['b',2]],columns=['col1','number'])
df2 = pd.DataFrame([['c',3,'Lion'],['d',4,'tiger']],columns=['letter','number','animal'])
```

```python
pd.concat([df1,df2],axis=0)
```

```python
pd.concat([df1,df2],axis=0,ignore_index=True)
```

```python
pd.concat([df1,df2],axis=1)
```

# Merging DataFrame

```python
d1 = pd.DataFrame({
    'city': ['delhi','lucknow','agra','kanpur'],
    'temperature': [32,45,30,40]
})
```

```python
d2 = pd.DataFrame({
    'city': ['lucknow','agra','kanpur'],
    'humidity': [30,43,50]
})
```

```python
d1
```

```python
d2
```

```python
df = pd.merge(d1,d2,on='city')
```

```python
df
```

```python
pd.merge(d1,d2,on='city',how='outer')
```

```python
pd.merge(d1,d2,on='city',how='left')
```
